```python
import numpy as np
from matplotlib import pyplot as plt
```

## test interpolation

```python
from scipy.interpolate import CloughTocher2DInterpolator
```

```python
# domain boundaries
Xmax = 4.5e-3
Ymax = Xmax

Rmax = Xmax
```

```python
# coordinates
nx = 51
ny = 53
xs = np.linspace(-Xmax, Xmax, nx)
ys = np.linspace(-Ymax, Ymax, ny)
Xs, Ys = np.meshgrid(xs, ys)

# nr = (nx+1)//2
nr = 201
nphi = 200
rs = np.linspace(0, Rmax, nr)
phis = np.linspace(0, 2*np.pi, nphi, endpoint=False)
Rs, Phis = np.meshgrid(rs, phis)
```

```python
w=1e-3
x0 = -0.5e-3 #0.1*Xmax
y0 = 1e-3
myfield4 = np.exp(-((Xs - x0)**2 + (Ys - y0)**2)/w**2)
```

```python
# interp = RegularGridInterpolator((ys,xs), myfield4, method='linear')
interp = CloughTocher2DInterpolator(list(zip(Xs.flat,Ys.flat)), myfield4.flat)
```

```python
myfield4_interp = interp(Rs*np.cos(Phis), Rs*np.sin(Phis))
```

```python
plt.pcolormesh(Xs, Ys, myfield4)
```

```python
plt.pcolormesh(Rs*np.cos(Phis), Rs*np.sin(Phis), myfield4_interp)
```
